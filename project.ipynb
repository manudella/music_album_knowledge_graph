### **3. Include a small ontology**

Using RDFlib, add classes such as Person, MusicAlbum, and so on to your KG. Add the domain and range for each property you use.

This will be needed later for inferring the types of the entities of your KG.

```python
#classes
#person
person=URIRef("http://example.org/Person")
g.add((person, RDF.type, OWL.Class))
g.add((person, RDFS.subClassOf, OWL.Thing))

#musicartis
musicartist=URIRef("http://example.org/MusicArtist")
g.add((musicartist, RDF.type, OWL.Class))
g.add((musicartist, RDFS.subClassOf, OWL.Thing))

#musicgroup
musicgroup=URIRef("http://example.org/MusicGroup")
g.add((musicgroup, RDF.type, OWL.Class))
g.add((musicgroup, RDFS.subClassOf, musicartist))

#musicsolist
musicsolist=URIRef("http://example.org/MusicSolist")
g.add((musicsolist, RDF.type, OWL.Class))
g.add((musicsolist, RDFS.subClassOf, musicartist))
g.add((musicsolist, RDFS.subClassOf, person))

#musicalbum
musicalbum=URIRef("http://example.org/MusicAlbum")
g.add((musicalbum, RDF.type, OWL.Class))
g.add((musicalbum, RDFS.subClassOf, OWL.Thing))

#musicgenre
musicgenre=URIRef("http://example.org/MusicGenre")
g.add((musicgenre, RDF.type, OWL.Class))
g.add((musicgenre, RDFS.subClassOf, OWL.Thing))

#musicproducer
musicproducer=URIRef("http://example.org/MusicGenre")
g.add((musicgenre, RDF.type, OWL.Class))
g.add((musicgenre, RDFS.subClassOf, OWL.Thing))

#property
#dbp.producer
g.add((dbp.producer, RDF.type, OWL.ObjectProperty))
g.add((dbp.producer, RDFS.domain, musicalbum))
g.add((dbp.producer, RDFS.range, musicproducer))

#dbp.artist
g.add((dbp.artist, RDF.type, OWL.ObjectProperty))
g.add((dbp.artist, RDFS.domain, musicalbum))
g.add((dbp.artist, RDFS.range, musicartist))

#dbo.genre
g.add((dbo.genre, RDF.type, OWL.ObjectProperty))
g.add((dbo.genre, RDFS.domain, musicalbum))
g.add((dbo.genre, RDFS.range, musicgenre))




print(g.serialize())

# Your solution ...
```


### **4. Genres taxonomy**

The first dataframe has 'Genre' and 'Subgenre' columns.

Extract a genres taxonomy from the dataset and include it in your RDF KG. This will be needed later for inferring genres from subgenres.

E.g.: add triples to the graph like:

- `[subgenre] rdfs:subClassOf [genre]`.
- `[album] rdf:type [genre]`.

This should be done for all genres and subgenres (with an iteration, not manually).

```python
gen = ["Metal"] #create the list for all genres
triples3=[]

for x in range(len(df3)): #separate and preprocess strings before checking for relations
    
    if df3['Genre_x'][x] != "nan": 
        if "," in str(df3['Genre_x'][x]):
            genx = str(df3['Genre_x'][x]).split(',')
            for x in range(len(genx)):
                if x > 0:
                    sgen = str(genx[x]).replace(" ","_").replace('"',"").lstrip("_")
                    if sgen not in gen:
                        gen.append(sgen)
                else:
                    sgen = str(genx[x]).replace(" ","_").replace('"',"")
                    if sgen not in gen:
                        gen.append(sgen)
                    
for x in range(len(df3)):
    
    if df3['Genre_y'][x] != "nan":
        if "," in str(df3['Genre_y'][x]):
            genx = str(df3['Genre_y'][x]).split(',')
            for x in range(len(genx)):
                if x > 0:
                    sgen = str(genx[x]).replace(" ","_").replace('"',"").lstrip("_")
                    if sgen not in gen:
                        gen.append(sgen)
                else:
                    sgen = str(genx[x]).replace(" ","_").replace('"',"")
                    if sgen not in gen:
                        gen.append(sgen)
                    
for x in range(len(df3)):
    
    if df3['Subgenre'][x] != "nan":
        if "," in str(df3['Subgenre'][x]):
            genx = str(df3['Subgenre'][x]).split(',')
            for x in range(len(genx)):
                if x > 0:
                    sgen = str(genx[x]).replace(" ","_").replace('"',"").lstrip("_")
                    if sgen not in gen:
                        gen.append(sgen)
                else:
                    sgen = str(genx[x]).replace(" ","_").replace('"',"")
                    if sgen not in gen:
                        gen.append(sgen)


for genre in gen: #creating triples 
    for x in range(len(gen)):
        if genre in gen[x] and genre != gen[x] : #check if a string is contained in another one to determine subclasses
            trip = URIRef("http://example.org/")+gen[x], RDFS.subClassOf, URIRef("http://example.org/")+genre
            triples3.append(trip)
            
for triple in triples3: #adding triples
    g.add(triple) 
    
print(g.serialize())            
```
